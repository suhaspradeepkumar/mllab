```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Step 1: Input Data
data = np.array([115.3, 195.5, 120.5, 110.2, 90.4,
                 105.6, 110.9, 116.3, 122.3, 125.4])

print("Original Data:", data)

# Step 2: Basic Statistics
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)

print("\n--- Basic Statistics ---")
print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)

# Step 3: Mode (check manually)
values, counts = np.unique(data, return_counts=True)
if counts.max() > 1:
    mode = values[counts.argmax()]
else:
    mode = "No mode"

print("Mode:", mode)

# Step 4: Min-Max Normalization
# Formula: (x - min) / (max - min)
scaler_minmax = MinMaxScaler()
normalized = scaler_minmax.fit_transform(data.reshape(-1, 1))

print("\n--- Min-Max Normalization ---")
print(normalized.flatten())

# Step 5: Standardization (Z-score)
# Formula: (x - mean) / std_dev
scaler_standard = StandardScaler()
standardized = scaler_standard.fit_transform(data.reshape(-1, 1))

print("\n--- Standardization (Z-score) ---")
print(standardized.flatten())
```
