```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

# Step 1: Load Dataset (download zoo.data from UCI)
cols = ["animal_name","hair","feathers","eggs","milk","airborne",
        "aquatic","predator","toothed","backbone","breathes",
        "venomous","fins","legs","tail","domestic","catsize","type"]

df = pd.read_csv("zoo.data", names=cols)

# Step 2: Prepare Data
X = df.drop(["animal_name","type"], axis=1)  # features
y = df["type"]  # target

# Convert to binary (example: type 1 vs others)
y_binary = (y == 1).astype(int)

# Step 3: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_binary, test_size=0.3, random_state=42)

# Step 4: Train Decision Tree
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Step 5: Predictions
y_pred = model.predict(X_test)

# Step 6: Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Step 7: Classification Report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```
